```python
counts = df['type'].value_counts()

plt.figure(figsize=(4,4))
plt.pie(counts, labels=counts.index, autopct='%1.1f%%', startangle=140, wedgeprops={'width': 0.3})
plt.title("Distribution of Movies and TV Shows on Netflix")
plt.show()
```

```python
counts = df['rating'].value_counts()

# Create a donut chart
plt.figure(figsize=(6, 6))
plt.pie(counts, labels=counts.index, autopct='%1.1f%%', startangle=140, wedgeprops={'width': 0.6})
plt.title("Ratings on NETFLIX")
plt.show()
```

```python
df['release_year'] = df['release_year'].astype(str)

du = df['release_year'].value_counts().sort_index()
plt.figure(figsize=(15, 10))
sb.barplot(x=du.index, y=du.values)
plt.title("content released details over years")
plt.xticks(rotation=45)  
plt.show()
```
