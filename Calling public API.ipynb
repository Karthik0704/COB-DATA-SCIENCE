```python
import pandas as pd
import requests
```

# Here we are going to call out BINANCE API 

```python
response = requests.get('https://api.binance.com/api/v3/ticker/24hr')
```

# Storing data from API to CSV file named  bitcoin_data.csv

```python
if response.status_code == 200:
    data = response.json()
    
    df = pd.DataFrame(data)
    
    display(df)
    
 
    df.to_csv('bitcoin_data.csv', index=False)
    
    print('Data saved to bitcoin_data.csv')
else:
    print('Error:', response.status_code)
```
